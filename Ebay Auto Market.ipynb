```python
autos_major_brands = autos_major_brands[(autos_major_brands.registration_year < 2018)]
autos_major_brands = autos_major_brands[(autos_major_brands.registration_year > 1901)]
```

```python
autos_major_brands.price = [x.strip('$') for x in autos_major_brands.price]
```

```python
autos_major_brands.price = autos_major_brands.price.str.replace(',','')
autos_major_brands['price'] = pd.to_numeric(autos_major_brands['price'])
```

```python
autos_major_brands = autos_major_brands[(autos_major_brands.price < 500000)]
```

```python
autos_major_brands.info()
```

```python
brands_mean_price = autos_major_brands.groupby('brand').mean()[['price']].sort_values(by='brand', ascending=False)
```

```python
print(brands_mean_price)
```

```python
autos_major_brands.odometer = [x.strip('km') for x in autos_major_brands.odometer]
```

```python
autos_major_brands.odometer = autos_major_brands.odometer.str.replace(',','')
autos_major_brands['odometer'] = pd.to_numeric(autos_major_brands['odometer'])
```

```python
autos_major_brands.head()
```

```python
brands_mean_mileage = autos_major_brands.groupby('brand').mean()[['odometer']].round(2).sort_values(by='brand', ascending=False)

```

```python
brands_mean_mileage
```

```python
brands_mean_prices_mileage = pd.concat([brands_mean_mileage, brands_mean_price], axis=1)
```

```python
brands_mean_prices_mileage
```
